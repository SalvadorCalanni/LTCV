This script shold be used to measure the real contrast of the looming stimuli in the arena. You should already have an idea of how your setup is working with the other script before using this one (Visual_looming_stimuly_with_variable_contrast).
You will also need a lux meter. Set it as in the middle of the area as indicated in the iumage.

```python
import turtle
import time
import timeit
import random
```

```python
def hacer_disco(color):
    disco = turtle.Turtle()
    disco.shape("circle")
    disco.color(color)
    return disco


def estimulo (color, diam_max, incremento, cronometrar = False):
    disco = hacer_disco(color)
    rep = 0
    while rep < 1:
        diametro = 0.1
        disco.st()
        inicio = time.time()
        while diametro < diam_max:
            diametro += incremento
            disco.shapesize(diametro, diametro)
        fin = time.time()
        tot = round(fin - inicio,3)
        if cronometrar:
            print ("Time of disk expansion", tot, "sec")        
        time.sleep(0.5)
#        disco.ht()
        rep += 1
    print(color)


    
def sesion(lista, diam_max, incremento, cronometrar = False):
    
    archivo = open("Contrast_list", "w")
    print("Contrast_list", file=archivo)
    print(" ", file=archivo)
    print( "Gray_code(Hex)", "Luminicence(lux)", file=archivo)

#    input ("Press enter to start timing (ideally, at the same time you start recording)")
#    inicio_sesion = time.time() 
    res = input ("Type here background illuminance. Then, press enter")
    print( "background", res, file=archivo)

    while len(lista)>0:
        

        color = lista[0]
        estimulo (color, diam_max, incremento, cronometrar)
        lista.remove(color)
        
        res = input ("Type here the measured illuminance. Then, press enter")
        print( color, res, file=archivo)
        
        print(len(lista), "stimuly left to be measured")
        input("press enter to measure the next stimuly")
        
    if len(lista) == 0:
        archivo.close()
        print("Nice! a file named 'Contrast_list' was created with your data")
```
